## fullname
1. Read 100 users first and last name
2. Print each users full name, and give them an id.
3. Each id should be 5 digits long. Every id is previous id + 1. For ex: 00001, 00002, 00003, ..., 00100


Notice how I broke this code into different "cells" i.e. code blocks. All the code I run will stay in Jupyter memory i.e. Jupyter will remember you wrote them, and you don't have to keep re running this code everytime you make minor changes in other parts of the code. How does this help? Maybe you want to keep getting the names of 100 people many many times. Why would you want to waste time & RAM to process old code that hasn't changed? 

```python
def getFull(a,b):
  res = str(a) + ' ' + str(b)
  return res;

def getFixedLengthId(numId):
  print('Num is was ', numId)
  return format(numId, '05d')

```

```python
i = 0
while (i < 100):
  print('Reading name for user', i)
  first = input('First Name?: ')
  last = input('Last name?: ')
  full = getFull(first,last)
  print(full)
  print(getFixedLengthId(i))
  i += 1
  print('-'*30)
```
